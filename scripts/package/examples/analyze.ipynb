```python
from analyze_video.config import (
    CUTS_PATH,
    CUTS_DESCRIPTION_PATH,
    OLLAMA_URL,
    PERCENT_FRAMES_TO_SAMPLE,
)
from analyze_video.utils import *
import re
import os
```

```python
# Sample frames according to activity in video

for cut in os.listdir(CUTS_PATH):
    cut_folder_path = os.path.join(CUTS_DESCRIPTION_PATH, cut.split(".")[0])
    print(f"cuts folder path : {cut_folder_path}")
    # Example usage

    video_cut_path = os.path.join(CUTS_PATH, cut)
    print(cut)

    activity = get_activity_per_frame(video_cut_path)
    activity = smooth_signal_gaussian(activity)
    adaptive_samples = adaptive_frame_sampling(activity=activity)

    if os.path.isdir(cut_folder_path):
        shutil.rmtree(cut_folder_path)
        print(f"Nuked {cut_folder_path}")
    os.mkdir(os.path.join(cut_folder_path))
    plot_and_save_cut_activity(
        path=os.path.join(cut_folder_path, f"{cut}_activity_plot.png"),
        activity=activity,
        sampled_indices=adaptive_samples,
    )

    save_frames_from_video(video_cut_path, adaptive_samples, cut_folder_path)
```

```python
def make_summary():
    cut_folders = os.listdir(CUTS_DESCRIPTION_PATH)
    cut_folders = sorted(cut_folders, key=lambda x: int(re.search(r"\d+", x).group()))

    for cut_folder in os.listdir(CUTS_DESCRIPTION_PATH):
        print(cut_folder)
        cut_folder_full_path = os.path.join(CUTS_DESCRIPTION_PATH, cut_folder)
        print(cut_folder_full_path)

        folder_descriptions = []

        # Get all image files and sort them by numeric suffix
        images = [
            f for f in os.listdir(cut_folder_full_path) if f.lower().endswith(".jpg")
        ]
        print(images)
        images = sorted(images, key=lambda x: int(re.search(r"\d+", x).group()))

        for image in images:
            image_path = os.path.join(cut_folder_full_path, image)

            # Display image
            img = mpimg.imread(image_path)
            plt.imshow(img)
            plt.axis("off")
            plt.show()

            # Get description
            description = describe_image_with_llava(image_path)
            print(description)
            folder_descriptions.append(description)
        break


make_summary()
```
