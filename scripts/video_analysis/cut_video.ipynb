```python
from scenedetect import VideoManager, SceneManager
from scenedetect.detectors import ContentDetector
import subprocess
```

```python
video_path = "/home/tyler/python/agents/video_inputs/sunlight.mp4"
```

```python
video_manager = VideoManager([video_path])
scene_manager = SceneManager()
scene_manager.add_detector(ContentDetector(threshold=60.0))
```

```python
video_manager.set_downscale_factor()  # for speed
video_manager.start()


scene_manager.detect_scenes(frame_source=video_manager)
```

```python
scene_list = scene_manager.get_scene_list()
```

```python
for i, scene in enumerate(scene_list):
    start, end = scene

    print(f"Scene {i + 1}: Start {start.get_timecode()} - End {end.get_timecode()}")
```

```python
cmd = [
    "scenedetect",
    "-i",
    video_path,
    "detect-content",
    "-t",
    "60",
    "split-video",
    "--output",
    "/home/tyler/python/agents/cuts",
]

result = subprocess.run(cmd, capture_output=True, text=True)
print(result.stdout)  # command output
print(result.stderr)
```
